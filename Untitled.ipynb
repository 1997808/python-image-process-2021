```python
from skimage import io, color, data
img = data.astronaut()
gray = color.rgb2gray(img)
io.imshow(gray);
```

```python
from skimage import io, color, data
img = data.astronaut()
img_hsv = color.rgb2hsv(img)
io.imshow(img_hsv);
```

```python
import numpy as np
from skimage import io, draw
img = np.zeros((100, 100), dtype = np.uint8)
x, y = draw.disk(30, 30)
img[x, y] = 1
x1, y1 = draw.ellipse(60, 60, 10, 20)
img[x1, y1] = 1
io.imshow(img);
```
