```python
# 3 biến: đường huyết, di truyền, tuổi -> có bị tiểu đường hay không
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Hàm sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Load data từ file csv
data = pd.read_csv('data/dataset_diabetes4.csv').values
N, d = data.shape
x = data[:, 0:d-1].reshape(-1, d-1)
y = data[:, 3].reshape(-1, 1)
#print(x)

# Thêm cột 1 vào dữ liệu x
x = np.hstack((np.ones((N, 1)), x))
w = np.array([0.,0.1,0.1,0.1]).reshape(-1,1)

# Số lần lặp bước 2
numOfIteration = 1000
cost = np.zeros((numOfIteration,1))
learning_rate = 0.0001

for i in range(1, numOfIteration):    
    # Tính giá trị dự đoán
    y_predict = sigmoid(np.dot(x, w))
    cost[i] = -np.sum(np.multiply(y, np.log(y_predict)) + np.multiply(1-y, np.log(1-y_predict)))
    # Gradient descent
    w = w - learning_rate * np.dot(x.T, y_predict-y)
    #print(cost[i])

#Giá trị thực tế:
x1=100
x2=0.5
x3=50
#Giá trị dự đoán
y_p= 1 / (1 + np.exp(-(w[0]+w[1]*x1+w[2]*x2+w[3]*x3)))
print(y_p)
if y_p<0.5:
    print("Khả năng ít bị tiểu đường: ",y_p)
else:
    print("Khản năng cao bị tiểu đường:",y_p)
```

```python
#Nhập dữ liệu
# 3 biến: đường huyết, di truyền, tuổi -> có bị tiểu đường hay không
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras

# Load data từ file csv
data = pd.read_csv('data/dataset_diabetes4.csv').values
N, d = data.shape
X = data[:, 0:d-1].reshape(-1, d-1)
Y = data[:, 3].reshape(-1, 1)
print(X,Y)

#Tạo dữ liệu train và test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, shuffle = True)
X_test, X_validate, y_test, y_validate = train_test_split(X_test, y_test, test_size = 0.2, shuffle = True)
print('X_train shape: ', X_train.shape)
print('y_train shape: ', y_train.shape)
print('X_test shape: ', X_test.shape)
print('y_test shape: ', y_test.shape)
print('X_validate shape: ', X_validate.shape)
print('y_validate shape: ', y_validate.shape)
print(X_train, y_train)

#Tạo mô hình mạng
from keras.models import Sequential
from keras.layers import Dense

def get_model():    
    model = Sequential([
        Dense(3, input_shape = (3,), activation = 'sigmoid'), #3 neurons, Input Layer
        Dense(2, activation = 'sigmoid'),                      #2  neurons, Hidden Layer
        Dense(1)                                            #Output Layer
    ])                                                      #'sigmoid' activation
    model.compile(
        loss='mse',                                         #Trained using Mean square error loss (Cost function) 
        optimizer='adam' ,                                   #Hàm tối ưu để tìm cực trị
        metrics=['accuracy']
    )    
    return model

model = get_model()
model.summary()

#Finally training the model-->
history = model.fit(X_train, y_train,validation_data = (X_test, y_test),epochs = 1000,batch_size=3)

model.get_weights()
# evaluate the keras model
_,accuracy = model.evaluate(X, Y, verbose=0)
print('Accuracy: %.2f' % (accuracy*100))

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

#Test
X_test=[[200,0.8,40]]
y_predicted = model.predict(X_test)
print(y_predicted)
```
