```python
import matplotlib.pylab as pylab
from PIL import Image, ImageFont, ImageDraw, ImageOps
import numpy as np
import matplotlib.pylab as plt

im = Image.open("data/car.jpg")
im = ImageOps.grayscale(im)

threshold1=50
threshold2=100
threshold3=150

im1 = im.point(lambda x: threshold2 > x > threshold1)
im2 = im.point(lambda x: x > threshold2)
im3 = im.point(lambda x: x > threshold3)

fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0, 0].imshow(im)
ax[0, 1].imshow(im1)
ax[1, 0].imshow(im2)
ax[1, 1].imshow(im3)
```
