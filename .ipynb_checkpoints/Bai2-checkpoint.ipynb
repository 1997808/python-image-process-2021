```python
####### Cách 2: dùng phương pháp cân bằng tổ chức đồ (histogram equalization)
###Đọc và hiển thị ảnh gốc
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
image = cv2.imread("Saved Pictures/low_contrast.jpg") 
plt.imshow(image,cmap='gray')
#################################################################################
###Hiển thị từng kênh riêng biệt
b = image.copy()
# set green and red channels to 0
b[:, :, 1] = 0
b[:, :, 2] = 0


g = image.copy()
# set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0

r = image.copy()
# set blue and green channels to 0
r[:, :, 0] = 0
r[:, :, 1] = 0


# RGB - Blue
cv2.imshow('B-RGB', b)

# RGB - Green
cv2.imshow('G-RGB', g)

# RGB - Red
cv2.imshow('R-RGB', r)

cv2.waitKey(0)
#######################################################################################
###Cân bằng tổ chức đồ mỗi kênh riêng biệt
#Can bang luoc do xam cho kenh red
def hist_equalize(r):
    # 1. calcalate hist
    hist = cv2.calcHist([r],channels=[0],mask=None, #fullimage
                   histSize=[256],#full scale
                    ranges=[0,256]
                   )
    #2. normalize Hist
    h,w = r.shape[:2]
    hist=hist/(h*w)
    #3.calculate CDF
    cdf = np.cumsum(hist)
    s_k = (255*cdf-0.5).astype("uint8")
    return s_k
#Ve anh goc va anh luoc do xam
def plot_img_and_hist(r):
    hist = cv2.calcHist([r],channels=[0],mask=None, #fullimage
                   histSize=[256],#full scale
                    ranges=[0,256]
    )
    plt.subplot(1,2,1)
    plt.imshow(r)
    plt.subplot(1,2,2)
    plt.plot(hist,'b')
    plt.legend(('histogram'),loc='upper left')
#Can bang xam cho anh goc
s_k = hist_equalize(r) #muc xam moi
#Ham chuyen doi anh goc sang anh moi ung voi muc xam sau can bang
equalized_img = cv2.LUT(r , s_k)
#Ve anh sau can bang va luoc do xam moi
plot_img_and_hist(equalized_img)
###########################################################################
#Can bang luoc do xam cho kenh green
def hist_equalize(g):
    # 1. calcalate hist
    hist = cv2.calcHist([g],channels=[0],mask=None, #fullimage
                   histSize=[256],#full scale
                    ranges=[0,256]
                   )
    #2. normalize Hist
    h,w = g.shape[:2]
    hist=hist/(h*w)
    #3.calculate CDF
    cdf = np.cumsum(hist)
    s_k = (255*cdf-0.5).astype("uint8")
    return s_k
#Ve anh goc va anh luoc do xam
def plot_img_and_hist(g):
    hist = cv2.calcHist([g],channels=[0],mask=None, #fullimage
                   histSize=[256],#full scale
                    ranges=[0,256]
    )
    plt.subplot(1,2,1)
    plt.imshow(g)
    plt.subplot(1,2,2)
    plt.plot(hist,'b')
    plt.legend(('histogram'),loc='upper left')
#Can bang xam cho anh goc
s_k = hist_equalize(g) #muc xam moi
#Ham chuyen doi anh goc sang anh moi ung voi muc xam sau can bang
equalized_img = cv2.LUT(g , s_k)
#Ve anh sau can bang va luoc do xam moi
plot_img_and_hist(equalized_img)
########################################################################
#Can bang luoc do xam cho kenh blue
def hist_equalize(b):
    # 1. calcalate hist
    hist = cv2.calcHist([b],channels=[0],mask=None, #fullimage
                   histSize=[256],#full scale
                    ranges=[0,256]
                   )
    #2. normalize Hist
    h,w = b.shape[:2]
    hist=hist/(h*w)
    #3.calculate CDF
    cdf = np.cumsum(hist)
    s_k = (255*cdf-0.5).astype("uint8")
    return s_k
#Ve anh goc va anh luoc do xam
def plot_img_and_hist(b):
    hist = cv2.calcHist([b],channels=[0],mask=None, #fullimage
                   histSize=[256],#full scale
                    ranges=[0,256]
    )
    plt.subplot(1,2,1)
    plt.imshow(b)
    plt.subplot(1,2,2)
    plt.plot(hist,'b')
    plt.legend(('histogram'),loc='upper left')
#Can bang xam cho anh goc
s_k = hist_equalize(b) #muc xam moi
#Ham chuyen doi anh goc sang anh moi ung voi muc xam sau can bang
equalized_img = cv2.LUT(b , s_k)
#Ve anh sau can bang va luoc do xam moi
plot_img_and_hist(equalized_img)
##############################################
##Can bang luoc do xam cho ca anh 
def hist_equalize(image):
    # 1. calcalate hist
    hist = cv2.calcHist([image],channels=[0],mask=None, #fullimage
                   histSize=[256],#full scale
                    ranges=[0,256]
                   )
    #2. normalize Hist
    h,w = image.shape[:2]
    hist=hist/(h*w)
    #3.calculate CDF
    cdf = np.cumsum(hist)
    s_k = (255*cdf-0.5).astype("uint8")
    return s_k
#Ve anh goc va anh luoc do xam
def plot_img_and_hist(image):
    hist = cv2.calcHist([image],channels=[0],mask=None, #fullimage
                   histSize=[256],#full scale
                    ranges=[0,256]
    )
    plt.subplot(1,2,1)
    plt.imshow(image)
    plt.subplot(1,2,2)
    plt.plot(hist,'b')
    plt.legend(('histogram'),loc='upper left')
#Can bang xam cho anh goc
s_k = hist_equalize(image) #muc xam moi
#Ham chuyen doi anh goc sang anh moi ung voi muc xam sau can bang
equalized_img = cv2.LUT(image , s_k)
#Ve anh sau can bang va luoc do xam moi
plot_img_and_hist(equalized_img)
```
