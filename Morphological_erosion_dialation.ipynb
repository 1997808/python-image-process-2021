```python
from skimage.io import imread
from skimage.color import rgb2gray
import matplotlib.pylab as pylab
from skimage.morphology import binary_erosion, rectangle, binary_dilation, disk
def plot_image(image, title=''):
    pylab.title(title, size=20), pylab.imshow(image)
    pylab.axis('off')
im = rgb2gray(imread('data/car.jpg'))
pylab.gray()

pylab.figure(figsize=(20,10))
pylab.subplot(1,3,1), plot_image(im, 'original')
im1 = binary_erosion(im, rectangle(1, 1))
pylab.subplot(1,3,2), plot_image(im, 'erosion with rectangle size (1,5)')
im1 = binary_erosion(im, rectangle(1, 10))
pylab.subplot(1,3,3), plot_image(im, 'erosion with rectangle size (1,15)')
pylab.show()

pylab.figure(figsize=(20,10))
pylab.subplot(1,3,1), plot_image(im, 'original')
for d in range(1,3):
    pylab.subplot(1,3,d+1)
    im1 = binary_dilation(im,disk(7*d))
    plot_image(im1)
```
